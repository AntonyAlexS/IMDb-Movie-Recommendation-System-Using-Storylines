```python
import pandas as pd
data=pd.read_csv('/content/Movie_data.csv')
data.head()
```

```python
#Remove the numbers from Movie_Title
data['Movie_Title'] = data['Movie_Title'].str.replace(r'^\d+\.\s*', '', regex=True)
```

```python
#Drop index column
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
#Remove whitespaces
def remove_whitespace(text):
  return " ".join(text.split())
data['Storyline2']=data['Storyline'].apply(remove_whitespace)
```

```python
#Convert to lower
def lower(text):
  return text.lower()
data['Storyline2']=data['Storyline2'].apply(lower)
```

```python
#Tokenization
from nltk import word_tokenize
data['Storyline2']=data['Storyline2'].apply(word_tokenize)
```

```python
#Remove stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))

en_stopwords=stopwords.words('english')
def remove_stopwords(text):
  result=[]
  for token in text:
    if token not in en_stopwords:
      result.append(token)
  return result
data['Storyline2']=data['Storyline2'].apply(remove_stopwords)
```

```python
#Remove punctuations
from nltk.tokenize import RegexpTokenizer

def remove_punctations(text):
  tokenizer=RegexpTokenizer(r"\w+")
  text=tokenizer.tokenize(' '.join(text))
  return text
data['Storyline2']=data['Storyline2'].apply(remove_punctations)
```

```python
#Join the words
data['Storyline2'] = data['Storyline2'].apply(lambda x: ' '.join(x))
```

```python
#Converting the text data to feature vectors
from sklearn.feature_extraction.text import TfidfVectorizer # Import the TfidfVectorizer class
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(data['Storyline2'])
print(feature_vectors)
```

```python
%%writefile app2.py
import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import string
from nltk.corpus import stopwords
import nltk
import contractions

nltk.download('stopwords')

# Load CSV
df = pd.read_csv(r'/content/Movie_data.csv')
df['Movie_Title'] = df['Movie_Title'].str.replace(r'^\d+\.\s*', '', regex=True)
df.drop(columns=['Unnamed: 0'], inplace=True)

# Preprocess
def clean_text(text):
    text = " ".join(text.split()) #Remove Extra Spaces
    text = text.lower()
    text = contractions.fix(text) #Handle contractions
    text = text.translate(str.maketrans('', '', string.punctuation))
    stop_words = stopwords.words('english')
    tokens = [word for word in text.split() if word not in stop_words]
    return ' '.join(tokens)

df['Clean_Storyline'] = df['Storyline'].apply(clean_text)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['Clean_Storyline'])

# Streamlit UI
st.title("🎬 Movie Recommender Based on Storyline")
user_input = st.text_area("Enter a movie storyline:")

if st.button("Recommend"):
    cleaned_input = clean_text(user_input)
    input_vec = vectorizer.transform([cleaned_input])
    similarity_scores = cosine_similarity(input_vec, tfidf_matrix).flatten()
    top5 = similarity_scores.argsort()[-5:][::-1]
    st.subheader("Top 5 Recommendations:")
    i=1
    for index in top5:
        st.write(f"**{i}.{df['Movie_Title'][index]}**")
        st.write(df['Storyline'][index])
        i=i+1
```
